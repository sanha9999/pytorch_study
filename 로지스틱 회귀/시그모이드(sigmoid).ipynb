```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x): # 시그모이드 함수 정의
    return 1/(1+np.exp(-x))
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y, 'g')
plt.plot([0,0], [1.0, 0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid function')
plt.show()
```

```python
# W값을 변화시키면서 경사도의 변화를 알아보자.

x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(0.5 * x)
y2 = sigmoid(x) # 원래 그래프(sigmoid)
y3 = sigmoid(2 * x)

plt.plot(x, y1, 'r', linestyle='--')
plt.plot(x, y2, 'g')
plt.plot(x, y3, 'b', linestyle='--')
plt.plot([0,0], [1.0, 0.0], ':')
plt.title('Sigmoid Fuction')
```

```python
# b값을 변화시키면서 그래프의 좌, 우 이동을 알 수 있다!!

x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(x+0.5)
y2 = sigmoid(x+1)
y3 = sigmoid(x+1.5)

plt.plot(x, y1, 'r', linestyle='--') # x + 0.5
plt.plot(x, y2, 'g') # x + 1
plt.plot(x, y3, 'b', linestyle='--') # x + 1.5
plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
```
