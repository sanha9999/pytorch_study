```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datasets/선형회귀_train.csv')
```

```python
df
```

```python
df.isna().sum()
```

```python
df['y'].describe()
```

```python
df['y'] = df['y'].fillna(49)
```

```python
df.isna().sum()
```

```python
x_target = torch.FloatTensor(df['x'].values)
```

```python
x_target = x_target.reshape([700, 1])
```

```python
y_target = torch.FloatTensor(df['y'].values)
```

```python
y_target = y_target.reshape([700, 1])
y_target
```

```python
model = nn.Linear(1, 1)
```

```python
import torch.optim as optim
optimizer = optim.SGD(model.parameters(), lr=1e-5)
epochs = 100000

for epoch in range(epochs+1):
    prediction = model(x_target)
    
    cost = F.mse_loss(prediction, y_target)
    
    
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    
    
    if epoch % 10000 == 0:
        print("epoch : {}/{}, cost : {}".format(epoch, epochs, cost.item()))
    
```
